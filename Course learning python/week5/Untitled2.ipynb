```python
# Upload 89
import os


def open_file(path, file_need_change_name):
    with open(path + '\\' + file_need_change_name, 'r', encoding="utf8") as file_html:
        html_file = file_html.readlines()
    return html_file


def get_the_new_name_for_the_files(html_file, rplace_name_with_the_original):
    i = 0
    is_chapter_found = True
    while is_chapter_found:
        if 'Chapter' in html_file[i]:
            temp = str(html_file[i])
            index = temp.index('Chapter')
            while not temp[index].isdecimal():
                index += 1
            while temp[index] != '\n':
                rplace_name_with_the_original += temp[index]
                index += 1
            is_chapter_found = False
        i += 1
    return rplace_name_with_the_original


def arrange_the_new_name_of_files(rplace_name_with_the_original):
    rplace_name_with_the_original = rplace_name_with_the_original.replace('</title>', '').replace(':', '')
    list_split_by_space = rplace_name_with_the_original.split()
    temp_one = list_split_by_space[0]
    temp_with_zero = temp_one.zfill(3)
    rplace_name_with_the_original = rplace_name_with_the_original.replace(temp_one, temp_with_zero)
    return rplace_name_with_the_original


def rename_files_func(path):
    for file_need_change_name in os.listdir(path):
        if file_need_change_name.endswith('.html'):
            rplace_name_with_the_original = ''
            html_file = open_file(path, file_need_change_name)
            rplace_name_with_the_original = get_the_new_name_for_the_files(html_file, rplace_name_with_the_original)
            rplace_name_with_the_original = arrange_the_new_name_of_files(rplace_name_with_the_original)
            old_name_html_file = path + file_need_change_name
            new_name_html_file = path + rplace_name_with_the_original + '.html'
            os.rename(old_name_html_file, new_name_html_file)


rename_files_func('C:\\Users\\Eylon\\Desktop\\python\\week5\\resources\\harry potter\\')
```
